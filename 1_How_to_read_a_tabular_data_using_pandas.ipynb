# How do I read a tabular data file into pandas ?

```python
## Tabular which looks like a table having rows and columns
```

```python
user_cols = ['user_id','age','gender','occupation','zip_code']
df = pd.read_table('http://bit.ly/movieusers', sep='|',header=None, names=user_cols)
df
```

```python
df.loc[:,'gender'].value_counts(normalize = True) 
```

```python
df.loc[:,'age'].value_counts(normalize = True) 
```

```python
df.groupby(by = ['age']).describe() 
```
