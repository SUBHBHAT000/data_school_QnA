# How to remove a column from a pandas data frame

```python
import pandas as pd 
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo
```

```python
# Deleting the column 'Colors Reported'
ufo.drop('Colors Reported', axis=1,inplace=True)
```

```python
ufo
```

```python
# Deleting the columns 'City' and 'State'
ufo.drop(['City','State'],axis=1,inplace=True)
ufo
```

```python
#Deleting rows 2 and 3
ufo.drop([0,1],axis=0,inplace=True)
ufo
```
