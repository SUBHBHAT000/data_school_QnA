# How to create dummy variables in pandas

```python
import pandas as pd 
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head() 
```

```python
# To create a dummy variable on the sex column 
# method one is to use the map method on the series 'Sex'
train['Sex_male'] = train.Sex.map({'male':1,'female':0})
```

```python
train.head()
```

```python
# Another method is little bit more work but more flexible
```

```python
pd.get_dummies(train.Sex) # This creates one column for every possible values
```

```python
# Generally speaking , if we have k possible values for a categorical variables, we use k-1 dummy 
# varables to represent it
# In other words we want to delete one columns.
# here we will use iloc method to keep one column
pd.get_dummies(train.Sex).iloc[:,1:] 
```

```python
# Now since we need to add these columns back to the original data frame we need to add an indentifier to 
# identify the column from where it came by using the parameter prefix
pd.get_dummies(train.Sex, prefix='Sex').iloc[:,1:] 
```

```python
# Now let us create dummies on a  categorical column having more than one values
train.Embarked.value_counts() 
```

```python
pd.get_dummies(train.Embarked, prefix='Embarked').iloc[:,1:]
```

```python
# Now we have add thisi to the oiginal data frame
embarked_dummies = pd.get_dummies(train.Embarked, prefix='Embarked').iloc[:,1:]
```

```python
train = pd.concat([train,embarked_dummies], axis=1) 
train.head() 
```

```python
# Bonus Tip 
## How to pass a dataframe to pd.get_dummies
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
pd.get_dummies(train, columns=['Sex','Embarked']) 
```

```python
# Now we have to drop the additional columns, use the drop_first = True
pd.get_dummies(train, columns=['Sex','Embarked'], drop_first=True) 
```
