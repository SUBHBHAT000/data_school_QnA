# How to sort a pandas series or data frame

```python
import pandas as pd 
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies
```

```python
# Sorting the title series
movies.title.sort_values(ascending=False)
```

```python
# Using bracket notation
movies['title'].sort_values()
```

```python
# Sorting the entire movies data frame by the column title
movies.sort_values(by = 'title',ascending=True,inplace=True)
movies
```

```python
# Sorting the entire movies data frame by the column genre first and then title
movies.sort_values(by = ['genre','title'],ascending=True,inplace=True)
movies
```
